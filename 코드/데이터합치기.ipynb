```python
import pandas as pd
```

```python
train2021 = pd.read_csv('./데이터/train_subway21.csv',encoding='cp949')
train2022 = pd.read_csv('./데이터/train_subway22.csv',encoding='cp949')
train2023 = pd.read_csv('./데이터/train_subway23.csv',encoding='cp949')
```

```python
train2021 = train2021.drop(columns=['Unnamed: 0'])
train2022 = train2022.drop(columns=['Unnamed: 0'])
train2023 = train2023.drop(columns=['Unnamed: 0'])
```

```python
train2021.columns = [col.replace('train_subway21.', '') for col in train2021.columns]
train2022.columns = [col.replace('train_subway22.', '') for col in train2022.columns]
train2023.columns = [col.replace('train_subway23.', '') for col in train2023.columns]
```

```python
# 컬럼명 변경 딕셔너리
col_map = {
    'tm': '시간',
    'line': '호선',
    'station_number': '역번호',
    'station_name': '역명',
    'direction': '상하구분',
    'stn': 'AWS_지점코드',
    'ta': '기온',
    'wd': '풍향',
    'ws': '풍속',
    'rn_day': '일강수량',
    'rn_hr1': '시간강수량',
    'hm': '상대습도',
    'si': '일사량',
    'ta_chi': '체감온도',
    'congestion': '혼잡도'
}

# 함수로 처리
for df in [train2021,train2022,train2023]:
    df.rename(columns=col_map, inplace=True)
```

```python
train2021.to_csv('./데이터2/train_subway2021.csv', index=False, encoding='utf-8-sig')
train2022.to_csv('./데이터2/train_subway2022.csv', index=False, encoding='utf-8-sig')
train2023.to_csv('./데이터2/train_subway2023.csv', index=False, encoding='utf-8-sig')
```
